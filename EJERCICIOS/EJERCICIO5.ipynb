### Ejercicio 1

```python
# Declaramos 2 variables

nuevo_alumno = ''
alumnos = []
```

```python
# Escribe el ciclo while solicitado

while nuevo_alumno.lower() != 'alto':
    if nuevo_alumno:
        alumnos.append(nuevo_alumno)
    nuevo_alumno = input('Ingresa el nombre del nuevo alumno ')

# Otra opcion
nuevo_alumno = input('Ingresa el nombre del nuevo alumno ')
while nuevo_alumno.lower() != 'alto':
  alumnos.append(nuevo_alumno)
  nuevo_alumno = input('Ingresa el nombre del nuevo alumno ')
```
